<h2> Limpieza y preparación de los datos

```python
import pandas as pd
```

```python
with open('bookscraper/bookscraper/depasdata.json', 'r', encoding='utf-8') as file:
    json_content = file.read()

# Al ejecutar de nuevo el scrapper, añade una nueva lista de diccionarios a los datos previos, con esto las integro
json_content = json_content.replace("][", ",")

# Escribe el contenido actualizado en el archivo
with open('bookscraper/bookscraper/depasdata.json', 'w', encoding='utf-8') as file:
    file.write(json_content)
```

```python
#Repite el proceso para los inmuebles de tipo casa
with open('bookscraper/bookscraper/casasdata.json', 'r', encoding='utf-8') as file:
    json_content = file.read()

json_content = json_content.replace("][", ",")
with open('bookscraper/bookscraper/casasdata.json', 'w', encoding='utf-8') as file:
    file.write(json_content)
```

```python
#Abro ambos sets de datos en dataframes de Pandas, y los combino en uno solo
df = pd.read_json('bookscraper/bookscraper/depasdata.json')
casas = pd.read_json('bookscraper/bookscraper/casasdata.json')
props = pd.concat([df,casas], ignore_index=True)
```

```python
#Estos valores no especifican su precio
props[props['precio']<10]
```

```python
#y estos valores necesitan multiplicarse por 1000
props[props['precio']<1000]
```

```python
#Defino la función que eliminará los valores que no tienen presio
def eliminar_sin_precio(value):
    if value < 10:
        return value * 1000
    else:
        return value

props['precio'] = props['precio'].apply(eliminar_sin_precio)
```

```python
# Defino la función para multiplicar por 1000 si el valor es menor de 1000
def multiply_by_1000(value):
    if value < 1001:
        return value * 1000
    else:
        return value

props['precio'] = props['precio'].apply(multiply_by_1000)
```

```python
# Defino una función que determina si un inmueble es una casa o departamento y crea una nueva columna en el dataframe con esta información:
def casa_o_depa(titulo):
    if 'casa' in titulo.lower():
        return "casa"
    elif 'departamento' in titulo.lower():
        return "departamento"
    elif 'depa' in titulo.lower():
        return 'departamento'
    else:
        return 'no_aclara'
    
props['tipo_inmueble'] = props['titulo'].apply(casa_o_depa)
columna_tipo_inmueble = props.pop('tipo_inmueble')
props.insert(1, 'tipo_inmueble', columna_tipo_inmueble)
```

```python
#En la web, los datos de longitud y latitud están intercambiados, aquí los ordeno como corresponde
props = props.rename(columns={'longitud_x': 'latitud', 'latitud_y':'longitud'})
```

```python
#Calculo la media del tamaño para cada tipo de propiedad
media_tamano_depas = props[props['tipo_inmueble'] == 'departamento']['tamaño_m2'].mean()
media_tamano_casas = props[props['tipo_inmueble'] == 'casa']['tamaño_m2'].mean()
```

```python
#y los imputo
props.loc[(props['tipo_inmueble'] == 'departamento') & (props['tamaño_m2'].isnull()), 'tamaño_m2'] = media_tamano_depas
props.loc[(props['tipo_inmueble'] == 'casa') & (props['tamaño_m2'].isnull()), 'tamaño_m2'] = media_tamano_casas
```

```python
# y lo mismo con habitaciones y con baños

media_banio_depas = props[props['tipo_inmueble'] == 'departamento']['banios'].mean()
media_banio_casas = props[props['tipo_inmueble'] == 'casa']['banios'].mean()
media_habs_depas = props[props['tipo_inmueble'] == 'departamento']['habitaciones'].mean()
media_habs_casas = props[props['tipo_inmueble'] == 'casa']['habitaciones'].mean()
```

```python
props.loc[(props['tipo_inmueble'] == 'departamento') & (props['banios'].isnull()), 'banios'] = media_banio_depas
props.loc[(props['tipo_inmueble'] == 'casa') & (props['banios'].isnull()), 'banios'] = media_banio_casas
props.loc[(props['tipo_inmueble'] == 'departamento') & (props['habitaciones'].isnull()), 'habitaciones'] = media_habs_depas
props.loc[(props['tipo_inmueble'] == 'casa') & (props['habitaciones'].isnull()), 'habitaciones'] = media_habs_casas
```

```python
#Limpio la columna de descripción, para luego eliminar todos los registros de propiedades en anticresis, porque no muestran un valor real de venta
props['descripcion'] = props['descripcion'].apply(lambda lista: ' '.join(map(str, lista)))
anticresis = props[props['descripcion'].str.contains('anticresis', case=False)]
props.drop(anticresis.index, inplace=True)
```

```python
#Existen propiedades listadas con un tamaño irreal (por debajo de los 25 m2), esas son eliminadas
sin_tamanio = props[props['tamaño_m2']<25]
props.drop(sin_tamanio.index, inplace=True)
```

```python
#Después de las limpiezas, elimino posibles duplicados
props = props.drop_duplicates(ignore_index = True)
```

```python
#Agrego la columna de precio por metro cuadrado
props['precio_x_m2'] = props['precio']/props['tamaño_m2']
```

Me interesa crear una nueva columna en el dataframe que indique a qué distrito de la ciudad pertenece cada propiedad, en casi la totalidad de registros listados esta información se encuentra en la dirección o en la descripción. El código a continuación busca esta información dentro de los datos de cada propiedad y la agrega en una nueva columna del dataframe

```python
#Defino una lista con todos los distritos de la ciudad. Incluyo algunos aparentemente repetidos que pueden variar en su tildación para que no se escape ninguna palabra clave
distritos = [
    "selva alegre",
    "cayma",
    "umacollo",
    "cerro colorado",
    "characato",
    "chiguata", "caraveli", "caravelí", "mollendo", "islay", "camana", "camaná"
    "hunter",
    "bustamante",
    "la joya",
    "mariano melgar",
    "miraflores",
    "mollebaya",
    "paucarpata",
    "pocsi",
    "polobaya",
    "quequeña",
    "sabandía", "sabandia", "sachaca",
    "siguas",
    "tarucani",
    "socabaya",
    "tiabaya",
    "uchumayo",
    "vítor", "vitor",
    "yanahuara",
    "yarabamba", "majes", 
    "yura", 'mollendo', 'matarani'
]
```

```python
#Defino funciones que buscarán el nombre del distrito en la dirección y en la descripción, en ese orden
def buscar_distrito(direccion):
    for distrito in distritos:
        if distrito in direccion.lower():
            return distrito
    return pd.NA
```

```python
#Un distrito tiene el mismo nombre de la ciudad, así que estos datos son tratados con una función distinta
def cercado(df):
    if pd.isna(df['distrito']):
        if df['direccion'].lower() == 'arequipa':
            return 'arequipa'
    return df['distrito']
```

```python
def buscar_distrito_descr(df):
    if pd.isna(df['distrito']):
        descripcion_lower = df['descripcion'].lower()
        for distrito in distritos:
            if distrito in descripcion_lower:
                return distrito
    return df['distrito']
```

```python
#Esta segunda función busca palabras clave para el distrito central (Arequipa), que lleva el mismo nombe que la ciudad
def cercado_2(df):
    if pd.isna(df['distrito']):
        descripcion_lower = df['descripcion'].lower()
        cercados = ['cercado', 'centro', 'palomar', 'terminal', 'venezuela', 'unsa', 'vallecito', 'casona']
        for cercado in cercados:
            if cercado in descripcion_lower:
                return 'arequipa'
    return df['distrito']
```

```python
props['distrito'] = props['direccion'].apply(buscar_distrito)
props['distrito'] = props.apply(cercado, axis=1)
props['distrito'] = props.apply(buscar_distrito_descr, axis=1)
props['distrito'] = props.apply(cercado_2, axis=1)
```

```python
#Después de la búsqueda y reemplazo de distritos, relleno los que no pudieron encontrarse con "desconocido"
props.loc[props['distrito'].isna(), 'distrito'] = 'desconocido'
```
